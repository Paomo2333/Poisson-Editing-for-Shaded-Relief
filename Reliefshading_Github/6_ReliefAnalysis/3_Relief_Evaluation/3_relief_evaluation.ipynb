```python
"""
Refactored ice doline hillshade quality metrics (SRI / HSI / TGI)
and a unified 5x3 figure.

- Uses H300 / MDOW / Relief clips on the icedoline grid
- Uses the Poisson hillshade as the extended basemap
"""

from pathlib import Path

import os
import sys

import geopandas as gpd
import matplotlib as mpl
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
import numpy as np
import rasterio
from matplotlib.colors import BoundaryNorm, ListedColormap, TwoSlopeNorm
from matplotlib_scalebar.scalebar import ScaleBar
from rasterio.enums import Resampling
from rasterio.features import geometry_mask
from rasterio.warp import reproject
from rasterio.windows import from_bounds
from skimage.filters.rank import entropy
from skimage.morphology import disk
from skimage.util import img_as_ubyte
```

# 2. Helper functions

```python
# ---------------------------------------------------------------------
# 2. Helper functions
# ---------------------------------------------------------------------
def read_clip_hillshades(paths_dict):
    """Read the clipped H300 / MDOW / Relief rasters and convert nodata to NaN."""
    hs = {}
    for name, path in paths_dict.items():
        if not os.path.exists(path):
            print(f"[WARN] Missing {name} raster: {path}")
            continue
        with rasterio.open(path) as src:
            arr = src.read(1).astype("float32")
            nodata = src.nodata
        if nodata is not None:
            arr[arr == nodata] = np.nan
        arr[arr == -9999.0] = np.nan
        hs[name] = arr
        print(f"{name}: valid pixel count = {np.isfinite(arr).sum()}")
    return hs


def resample_to_grid(src_path, dst_shape, dst_transform, dst_crs, dst_nodata=DST_NODATA):
    """
    Resample the source raster to a target grid and normalize valid pixels to [0, 1].
    This helper is used to build H300_ext and MDOW_ext.
    """
    with rasterio.open(src_path) as src:
        src_crs = src.crs
        src_transform = src.transform
        src_nodata = src.nodata
        src_data = src.read(1).astype("float32")

    out = np.full(dst_shape, dst_nodata, dtype="float32")

    reproject(
        source=src_data,
        destination=out,
        src_transform=src_transform,
        src_crs=src_crs,
        src_nodata=src_nodata,
        dst_transform=dst_transform,
        dst_crs=dst_crs,
        dst_nodata=dst_nodata,
        resampling=Resampling.bilinear,
    )

    out[out == dst_nodata] = np.nan
    out[out == -9999.0] = np.nan

    valid = np.isfinite(out)
    if np.any(valid):
        vmin = np.nanmin(out[valid])
        vmax = np.nanmax(out[valid])
        if vmax > vmin:
            out = (out - vmin) / (vmax - vmin)
    return out


def compute_local_entropy(img_float, radius=ENTROPY_RADIUS):
    """Compute local entropy as the texture metric T and keep NaN outside valid areas."""
    selem = disk(radius)
    arr = img_float.copy().astype("float32")
    valid = np.isfinite(arr)
    arr[~valid] = 0.0
    arr_u8 = img_as_ubyte(np.clip(arr, 0.0, 1.0))
    T = entropy(arr_u8, footprint=selem).astype("float32")
    T[~valid] = np.nan
    return T


def safe_ratio(num, denom):
    return float(num) / float(denom) if denom > 0 else np.nan


def add_map_decoration(ax, title, pixel_size, x_left, x_right, y_bottom, y_top, icedoline_gdf):
    """Apply a consistent extent, title, scalebar, and north arrow to one map panel."""
    ax.set_xlim(x_left, x_right)
    ax.set_ylim(y_bottom, y_top)
    ax.axis("off")
    ax.set_title(title, fontsize=11)

    icedoline_gdf.boundary.plot(ax=ax, edgecolor="yellow", linewidth=1.5, linestyle="--")

    scalebar = ScaleBar(
        pixel_size,
        units="m",
        dimension="si-length",
        location="lower right",
        length_fraction=0.25,
        box_alpha=0.6,
        color="black",
        scale_loc="top",
        font_properties={"size": 9},
        fixed_value=100,
        fixed_units="m",
    )
    ax.add_artist(scalebar)
    add_north_arrow(ax, location="upper left", size=0.16)
```

# 3. Load the reference grid and IceDoline polygon

```python

# ---------------------------------------------------------------------
# 3. Load the reference grid and IceDoline polygon
# ---------------------------------------------------------------------
icedoline_gdf = gpd.read_file(shp_path)

with rasterio.open(target_files["H300"]) as src_ref:
    ref_crs       = src_ref.crs
    ref_transform = src_ref.transform
    ref_height    = src_ref.height
    ref_width     = src_ref.width

icedoline_ref = icedoline_gdf.to_crs(ref_crs)

print("Reference grid CRS:", ref_crs)
print("Reference grid shape (H, W):", ref_height, ref_width)

# Extent of the icedoline clip, used to build overlay masks
xmin_clip, ymin_clip, xmax_clip, ymax_clip = rasterio.transform.array_bounds(
    ref_height, ref_width, ref_transform
)

# Mask for the doline interior; all later metrics are evaluated inside this mask
mask_inside = geometry_mask(
    [geom for geom in icedoline_ref.geometry],
    out_shape=(ref_height, ref_width),
    transform=ref_transform,
    invert=True,
)
```

# 4. Load H300 / MDOW / Relief clips

```python

# ---------------------------------------------------------------------
# 4. Load H300 / MDOW / Relief clips
# ---------------------------------------------------------------------
hs_clip = read_clip_hillshades(target_files)
h300   = hs_clip["H300"]
mdow   = hs_clip["MDOW"]
relief = hs_clip["Relief"]
```

# 5. Load slope and aspect, then resample them to the icedoline grid

```python

# ---------------------------------------------------------------------
# 5. Load slope and aspect, then resample them to the icedoline grid
# ---------------------------------------------------------------------
with rasterio.open(slope_path) as src_slope:
    slope_crs       = src_slope.crs
    slope_transform = src_slope.transform
    slope_nodata    = src_slope.nodata
    slope_data      = src_slope.read(1).astype("float32")

slope_resampled = np.full((ref_height, ref_width), DST_NODATA, dtype="float32")

reproject(
    source=slope_data,
    destination=slope_resampled,
    src_transform=slope_transform,
    src_crs=slope_crs,
    src_nodata=slope_nodata,
    dst_transform=ref_transform,
    dst_crs=ref_crs,
    dst_nodata=DST_NODATA,
    resampling=Resampling.bilinear,
)

slope_resampled[slope_resampled == DST_NODATA] = np.nan
slope_resampled[slope_resampled == -9999.0] = np.nan
slope_deg = slope_resampled

with rasterio.open(aspect_path) as src_aspect:
    aspect_crs       = src_aspect.crs
    aspect_transform = src_aspect.transform
    aspect_nodata    = src_aspect.nodata
    aspect_data      = src_aspect.read(1).astype("float32")

aspect_resampled = np.full((ref_height, ref_width), DST_NODATA, dtype="float32")

reproject(
    source=aspect_data,
    destination=aspect_resampled,
    src_transform=aspect_transform,
    src_crs=aspect_crs,
    src_nodata=aspect_nodata,
    dst_transform=ref_transform,
    dst_crs=ref_crs,
    dst_nodata=DST_NODATA,
    resampling=Resampling.bilinear,
)

aspect_resampled[aspect_resampled == DST_NODATA] = np.nan
aspect_resampled[aspect_resampled == -9999.0] = np.nan
aspect_deg = aspect_resampled

# Keep only pixels inside the doline
slope_icedoline  = slope_deg.copy()
aspect_icedoline = aspect_deg.copy()
slope_icedoline[~mask_inside]  = np.nan
aspect_icedoline[~mask_inside] = np.nan

print("Valid slope pixels inside the doline:", np.isfinite(slope_icedoline).sum())
print("Valid aspect pixels inside the doline:", np.isfinite(aspect_icedoline).sum())
```

# 6. Build the extended Poisson basemap plus H300_ext / MDOW_ext

```python

# ---------------------------------------------------------------------
# 6. Build the extended Poisson basemap plus H300_ext / MDOW_ext
# ---------------------------------------------------------------------
with rasterio.open(ref_base_path) as src_base:
    base_crs       = src_base.crs
    base_transform = src_base.transform
    base_nodata    = src_base.nodata

    icedoline_base = icedoline_ref.to_crs(base_crs)

    minx, miny, maxx, maxy = icedoline_base.total_bounds
    minx_b = minx - BUFFER_DOLINE
    miny_b = miny - BUFFER_DOLINE
    maxx_b = maxx + BUFFER_DOLINE
    maxy_b = maxy + BUFFER_DOLINE

    win = from_bounds(minx_b, miny_b, maxx_b, maxy_b, base_transform)
    poisson_ext = src_base.read(1, window=win).astype("float32")
    poisson_ext_transform = src_base.window_transform(win)

if base_nodata is not None:
    poisson_ext[poisson_ext == base_nodata] = np.nan
poisson_ext[poisson_ext == -9999.0] = np.nan

ext_height, ext_width = poisson_ext.shape
xmin_ext, ymin_ext, xmax_ext, ymax_ext = rasterio.transform.array_bounds(
    ext_height, ext_width, poisson_ext_transform
)

# Inset the display extent slightly to avoid edge noise
x_left   = xmin_ext + BUFFER_VIEW
x_right  = xmax_ext - BUFFER_VIEW
y_bottom = ymin_ext + BUFFER_VIEW
y_top    = ymax_ext - BUFFER_VIEW

pixel_size = abs(poisson_ext_transform.a)

# Resample full-scene H300 / MDOW to the Poisson extent
h300_ext = resample_to_grid(
    h300_total_path,
    dst_shape=poisson_ext.shape,
    dst_transform=poisson_ext_transform,
    dst_crs=base_crs,
)
mdow_ext = resample_to_grid(
    mdow_total_path,
    dst_shape=poisson_ext.shape,
    dst_transform=poisson_ext_transform,
    dst_crs=base_crs,
)

print("poisson_ext shape:", poisson_ext.shape)
print("h300_ext / mdow_ext shape:", h300_ext.shape, mdow_ext.shape)
```

# 7. Illumination geometry: compute cos(i)

```python

# ---------------------------------------------------------------------
# 7. Illumination geometry: compute cos(i)
# ---------------------------------------------------------------------
az_sun_deg  = 300.0
alt_sun_deg = 45.0

az_sun  = np.deg2rad(az_sun_deg)
alt_sun = np.deg2rad(alt_sun_deg)

slope_rad  = np.deg2rad(slope_icedoline)
aspect_rad = np.deg2rad(aspect_icedoline)

cos_i = np.full_like(slope_icedoline, np.nan, dtype="float32")
valid_geom = np.isfinite(slope_rad) & np.isfinite(aspect_rad)
cos_i[valid_geom] = (
        np.sin(alt_sun) * np.cos(slope_rad[valid_geom]) +
        np.cos(alt_sun) * np.sin(slope_rad[valid_geom]) *
        np.cos(az_sun - aspect_rad[valid_geom])
)
```

# 8. Base masks and the brightness / front-facing / slope classes

```python

# ---------------------------------------------------------------------
# 8. Base masks and the brightness / front-facing / slope classes
# ---------------------------------------------------------------------
valid_base = (
        mask_inside &
        np.isfinite(h300) &
        np.isfinite(cos_i) &
        np.isfinite(slope_icedoline)
)

# Brightness classes used for the display panels
dark_mask_b   = valid_base & (h300 < DARK_THR)
bright_mask_b = valid_base & (h300 > BRIGHT_THR_B)
mid_mask_b    = valid_base & ~dark_mask_b & ~bright_mask_b

# illumination front/back
valid_slope_mask = valid_base
front_mask = valid_slope_mask & (cos_i >= COS_FRONT_THR) & (slope_icedoline > SLOPE_THR_DEG)
back_mask  = valid_slope_mask & (cos_i <  COS_FRONT_THR) & (slope_icedoline > SLOPE_THR_DEG)

# slope threshold mask
slope_thr_mask = valid_slope_mask & (slope_icedoline > SLOPE_THR_DEG)

# Classification arrays used in the second-row panels
cat_brightness = np.full(h300.shape, np.nan, dtype="float32")
cat_brightness[dark_mask_b]   = 0
cat_brightness[mid_mask_b]    = 1
cat_brightness[bright_mask_b] = 2

cat_light = np.full(h300.shape, np.nan, dtype="float32")
cat_light[front_mask] = 0
cat_light[back_mask]  = 1

cat_slope = np.full(h300.shape, np.nan, dtype="float32")
cat_slope[valid_slope_mask & ~slope_thr_mask] = 0
cat_slope[slope_thr_mask]                     = 1

# colormaps
cmap_brightness = ListedColormap(["navy", "goldenrod", "red"])
norm_brightness = BoundaryNorm([-0.5, 0.5, 1.5, 2.5], cmap_brightness.N)

cmap_light = ListedColormap(["orange", "cyan"])
norm_light = BoundaryNorm([-0.5, 0.5, 1.5], cmap_light.N)

cmap_slope = ListedColormap(["lightgray", "purple"])
norm_slope = BoundaryNorm([-0.5, 0.5, 1.5], cmap_slope.N)
```

# 9. SRI: Shadow Recovery Index

```python

# ---------------------------------------------------------------------
# 9. SRI：Shadow Recovery Index
# ---------------------------------------------------------------------
slope_mask   = slope_icedoline > SLOPE_THR_DEG
illum_shadow = cos_i < COS_SHADOW_THR
deep_shadow  = h300 < DARK_THR

M_shadow = valid_base & slope_mask & illum_shadow & deep_shadow
N_shadow = int(M_shadow.sum())
print("Shadow-mask pixel count M_shadow:", N_shadow)

H300_img   = h300
Relief_img = relief
MDOW_img   = mdow

valid_H300   = M_shadow & np.isfinite(H300_img)
valid_Relief = M_shadow & np.isfinite(Relief_img)
valid_MDOW   = M_shadow & np.isfinite(MDOW_img)

M_shadow_mid_H300   = valid_H300   & (H300_img   >= MID_LOW) & (H300_img   <= MID_HIGH)
M_shadow_mid_Relief = valid_Relief & (Relief_img >= MID_LOW) & (Relief_img <= MID_HIGH)
M_shadow_mid_MDOW   = valid_MDOW   & (MDOW_img   >= MID_LOW) & (MDOW_img   <= MID_HIGH)

N_mid_H300   = int(M_shadow_mid_H300.sum())
N_mid_Relief = int(M_shadow_mid_Relief.sum())
N_mid_MDOW   = int(M_shadow_mid_MDOW.sum())

SRI_H300   = safe_ratio(N_mid_H300,   N_shadow)
SRI_Relief = safe_ratio(N_mid_Relief, N_shadow)
SRI_MDOW   = safe_ratio(N_mid_MDOW,   N_shadow)

print(f"SRI_H300   = {SRI_H300:.3f}")
print(f"SRI_Relief = {SRI_Relief:.3f}")
print(f"SRI_MDOW   = {SRI_MDOW:.3f}")
```

# 10. HSI: Highlight Suppression Index

```python

# ---------------------------------------------------------------------
# 10. HSI：Highlight Suppression Index
# ---------------------------------------------------------------------
valid_common = (
        mask_inside &
        np.isfinite(h300) &
        np.isfinite(relief) &
        np.isfinite(mdow) &
        np.isfinite(cos_i) &
        np.isfinite(slope_icedoline)
)

front_mask_hsi = valid_common & (cos_i >= COS_FRONT_THR) & (slope_icedoline > SLOPE_THR_DEG)
bright_mask_h  = valid_common & (h300 > BRIGHT_THR_H)

M_high = front_mask_hsi & bright_mask_h
N_high = int(M_high.sum())
print("Highlight-mask pixel count M_high:", N_high)

mid_h300   = (h300   >= MID_LOW) & (h300   <= MID_HIGH)
mid_relief = (relief >= MID_LOW) & (relief <= MID_HIGH)
mid_mdow   = (mdow   >= MID_LOW) & (mdow   <= MID_HIGH)

M_high_mid_H300   = M_high & mid_h300
M_high_mid_Relief = M_high & mid_relief
M_high_mid_MDOW   = M_high & mid_mdow

n_high_mid_H300   = int(M_high_mid_H300.sum())
n_high_mid_Relief = int(M_high_mid_Relief.sum())
n_high_mid_MDOW   = int(M_high_mid_MDOW.sum())

HSI_H300   = safe_ratio(n_high_mid_H300,   N_high)
HSI_Relief = safe_ratio(n_high_mid_Relief, N_high)
HSI_MDOW   = safe_ratio(n_high_mid_MDOW,   N_high)

print(f"HSI_H300   = {HSI_H300:.3f}")
print(f"HSI_Relief = {HSI_Relief:.3f}")
print(f"HSI_MDOW   = {HSI_MDOW:.3f}")
```

# 11. TGI: Texture Gain Index (local entropy)

```python

# ---------------------------------------------------------------------
# 11. TGI: Texture Gain Index (local entropy)
# ---------------------------------------------------------------------
print("\nComputing local entropy for T_h300 / T_relief / T_mdow ...")
T_h300   = compute_local_entropy(h300,   radius=ENTROPY_RADIUS)
T_relief = compute_local_entropy(relief, radius=ENTROPY_RADIUS)
T_mdow   = compute_local_entropy(mdow,   radius=ENTROPY_RADIUS)

print("Valid pixels in T_h300:", np.isfinite(T_h300).sum())
print("Valid pixels in T_relief:", np.isfinite(T_relief).sum())
print("Valid pixels in T_mdow:", np.isfinite(T_mdow).sum())

# Shadow zone: compare only inside M_shadow_mid_Relief
mask_shadow_eval = M_shadow_mid_Relief & np.isfinite(T_h300) & np.isfinite(T_relief)
dT_shadow_Relief_vs_H300 = T_relief - T_h300
TGI_shadow_Relief_vs_H300 = float(np.nanmean(dT_shadow_Relief_vs_H300[mask_shadow_eval]))

mask_shadow_eval_mdow = mask_shadow_eval & np.isfinite(T_mdow)
dT_shadow_Relief_vs_MDOW = T_relief - T_mdow
TGI_shadow_Relief_vs_MDOW = float(np.nanmean(dT_shadow_Relief_vs_MDOW[mask_shadow_eval_mdow]))

# Highlight zone: compare Relief vs H300 only inside M_high_mid_Relief
mask_high_eval = M_high_mid_Relief & np.isfinite(T_h300) & np.isfinite(T_relief)
dT_high_Relief_vs_H300 = T_relief - T_h300
TGI_high_Relief_vs_H300 = float(np.nanmean(dT_high_Relief_vs_H300[mask_high_eval]))

print(f"TGI_shadow (Relief - H300) = {TGI_shadow_Relief_vs_H300:.4f}")
print(f"TGI_shadow (Relief - MDOW) = {TGI_shadow_Relief_vs_MDOW:.4f}")
print(f"TGI_high   (Relief - H300) = {TGI_high_Relief_vs_H300:.4f}")

# Prepare the Delta-T maps for the spatial panels
dT_shadow_map_H300 = np.full_like(T_relief, np.nan, dtype="float32")
dT_shadow_map_H300[mask_shadow_eval] = dT_shadow_Relief_vs_H300[mask_shadow_eval]

dT_shadow_map_MDOW = np.full_like(T_relief, np.nan, dtype="float32")
dT_shadow_map_MDOW[mask_shadow_eval_mdow] = dT_shadow_Relief_vs_MDOW[mask_shadow_eval_mdow]

dT_high_map = np.full_like(T_relief, np.nan, dtype="float32")
dT_high_map[mask_high_eval] = dT_high_Relief_vs_H300[mask_high_eval]

all_dT_vals = np.concatenate([
    dT_shadow_map_H300[np.isfinite(dT_shadow_map_H300)].ravel(),
    dT_shadow_map_MDOW[np.isfinite(dT_shadow_map_MDOW)].ravel(),
    dT_high_map[np.isfinite(dT_high_map)].ravel(),
])
if all_dT_vals.size > 0:
    vmax_abs = np.nanpercentile(np.abs(all_dT_vals), 99)
else:
    vmax_abs = 1.0
vmin_dT, vmax_dT = -vmax_abs, vmax_abs
norm_dT = TwoSlopeNorm(vmin=vmin_dT, vcenter=0.0, vmax=vmax_dT)
```

```python
# ---------------------------------------------------------------------
# 12. Unified 5x3 figure (aligned panels, refined scalebars/north arrows, upper-right labels)
# ---------------------------------------------------------------------
from matplotlib.patches import Patch
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import matplotlib.pyplot as plt
import numpy as np
import string  # Used to generate labels such as (a), (b), (c) ...
from matplotlib.colors import ListedColormap, BoundaryNorm

# Font settings
FONT_MIN = 11
FONT_TITLE = 13

# ==============================================================================
# 1. Helper function
# ==============================================================================
def add_map_decoration(ax, title, pixel_size, x_left, x_right, y_bottom, y_top, icedoline_gdf):
    """Apply a consistent extent, title, scalebar, and north arrow to one map panel."""
    ax.set_xlim(x_left, x_right)
    ax.set_ylim(y_bottom, y_top)
    ax.axis("off")

    # Draw titles with text so every panel shares the same vertical position
    if title:
        ax.text(
            0.5, 1.03,                  # Use the same y-position for every panel
            title,
            transform=ax.transAxes,
            ha="center",
            va="bottom",
            fontsize=FONT_TITLE,
        )

    # Doline boundary
    icedoline_gdf.boundary.plot(
        ax=ax, edgecolor="yellow", linewidth=1.5, linestyle="--"
    )

    # Scalebar
    scalebar = ScaleBar(
        pixel_size,
        units="m",
        dimension="si-length",
        location="lower right",
        length_fraction=0.25,
        box_alpha=0.6,
        color="black",
        scale_loc="top",
        font_properties={"size": 11},  # Keep the text size consistent
        fixed_value=100,
        fixed_units="m",
    )
    ax.add_artist(scalebar)



    # Keep the north arrow large enough to read
    add_north_arrow(ax, location="upper left", size=0.2)

# ==============================================================================
# 2. Main plotting workflow
# ==============================================================================

fig, axes = plt.subplots(5, 3, figsize=(11, 15.0))
axes = np.array(axes)

# ------------------------ Row 1 ------------------------
# H300
ax = axes[0, 0]
ax.imshow(np.ma.masked_invalid(h300_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
add_map_decoration(ax, "Hillshade (300°)", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

# MDOW
ax = axes[0, 1]
ax.imshow(np.ma.masked_invalid(mdow_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
add_map_decoration(ax, "MDOW Hillshade", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

# Relief
ax = axes[0, 2]
im_gray = ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
add_map_decoration(ax, "Shaded Relief", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

# ------------------------ Row 2 ------------------------
cat_brightness_plot = cat_brightness.copy()
cat_brightness_plot[~slope_thr_mask] = np.nan

# Slope
ax = axes[1, 0]
ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
ax.imshow(np.ma.masked_invalid(cat_slope), cmap=cmap_slope, norm=norm_slope, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.7)
add_map_decoration(ax, f"Slope threshold (> {SLOPE_THR_DEG}°)", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)
# Legend
handles_slope = [Patch(facecolor='purple', edgecolor='k', label=f"Slope > {SLOPE_THR_DEG}°")]
ax.legend(handles=handles_slope, loc='lower left', bbox_to_anchor=(0, 0), borderaxespad=0, fontsize=FONT_MIN, frameon=True).get_frame().set_alpha(0.85)

# Brightness
ax = axes[1, 1]
ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
ax.imshow(np.ma.masked_invalid(cat_brightness_plot), cmap=cmap_brightness, norm=norm_brightness, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.6)
add_map_decoration(ax, "Dark/Mid/Bright Area", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)
# Legend
handles_bright = [Patch(facecolor='navy', edgecolor='k', label="Dark"), Patch(facecolor='goldenrod', edgecolor='k', label="Mid-tone"), Patch(facecolor='red', edgecolor='k', label="Bright")]
ax.legend(handles=handles_bright, loc='lower left', bbox_to_anchor=(0, 0), borderaxespad=0, fontsize=10, frameon=True).get_frame().set_alpha(0.85)

# Light
ax = axes[1, 2]
ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
ax.imshow(np.ma.masked_invalid(cat_light), cmap=cmap_light, norm=norm_light, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.6)
add_map_decoration(ax, "Front-lit/Back-lit", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)
# Legend
handles_light = [Patch(facecolor='orange', edgecolor='k', label="Front-lit"), Patch(facecolor='cyan', edgecolor='k', label="Back-lit")]
ax.legend(handles=handles_light, loc='lower left', bbox_to_anchor=(0, 0), borderaxespad=0, fontsize=FONT_MIN, frameon=True).get_frame().set_alpha(0.85)

# ------------------------ Row 3: SRI ------------------------
shadow_arr = np.full(h300.shape, np.nan, dtype="float32"); shadow_arr[M_shadow] = 1.0
rec_h300_arr = np.full(h300.shape, np.nan, dtype="float32"); rec_h300_arr[M_shadow_mid_H300] = 1.0
rec_mdow_arr = np.full(h300.shape, np.nan, dtype="float32"); rec_mdow_arr[M_shadow_mid_MDOW] = 1.0
rec_relief_arr = np.full(h300.shape, np.nan, dtype="float32"); rec_relief_arr[M_shadow_mid_Relief] = 1.0

cmap_shadow = ListedColormap(["none", "navy"])
cmap_rec    = ListedColormap(["none", "limegreen"])
norm_mask2  = BoundaryNorm([-0.5, 0.5, 1.5], 2)
sri_vals = [SRI_H300, SRI_MDOW, SRI_Relief]
sri_arrs = [rec_h300_arr, rec_mdow_arr, rec_relief_arr]
sri_labels = ["SRI-H300", "SRI-MDOW", "SRI-Relief"]

for j in range(3):
    ax = axes[2, j]
    ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
    ax.imshow(np.ma.masked_invalid(shadow_arr), cmap=cmap_shadow, norm=norm_mask2, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.75)
    ax.imshow(np.ma.masked_invalid(sri_arrs[j]), cmap=cmap_rec, norm=norm_mask2, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.9)
    add_map_decoration(ax, "", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

    if j != 0:
        ax.text(0.02, 0.02, f"{sri_labels[j]}: {sri_vals[j]:.3f}",
                transform=ax.transAxes, ha="left", va="bottom", fontsize=FONT_MIN,
                bbox=dict(facecolor="white", alpha=0.75, edgecolor="none", pad=1.5))

handles_sri = [Patch(facecolor='navy', edgecolor='k', label="Shadow Area"), Patch(facecolor='limegreen', edgecolor='k', label="Recovered Area")]
axes[2, 0].legend(handles=handles_sri, loc='lower left', bbox_to_anchor=(0, 0), borderaxespad=0, fontsize=FONT_MIN, frameon=True).get_frame().set_alpha(0.85)

# ------------------------ Row 4: HSI ------------------------
highlight_arr = np.full(h300.shape, np.nan, dtype="float32"); highlight_arr[M_high] = 1.0
supp_h300_arr = np.full(h300.shape, np.nan, dtype="float32"); supp_h300_arr[M_high_mid_H300] = 1.0
supp_mdow_arr = np.full(h300.shape, np.nan, dtype="float32"); supp_mdow_arr[M_high_mid_MDOW] = 1.0
supp_relief_arr = np.full(h300.shape, np.nan, dtype="float32"); supp_relief_arr[M_high_mid_Relief] = 1.0

cmap_high = ListedColormap(["none", "darkred"])
cmap_supp = ListedColormap(["none", "deepskyblue"])
hsi_vals = [HSI_H300, HSI_MDOW, HSI_Relief]
hsi_arrs = [supp_h300_arr, supp_mdow_arr, supp_relief_arr]
hsi_labels = ["HSI-H", "HSI-MDOW", "HSI-SR"]

for j in range(3):
    ax = axes[3, j]
    ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
    ax.imshow(np.ma.masked_invalid(highlight_arr), cmap=cmap_high, norm=norm_mask2, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.75)
    ax.imshow(np.ma.masked_invalid(hsi_arrs[j]), cmap=cmap_supp, norm=norm_mask2, extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.9)
    add_map_decoration(ax, "", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

    if j != 0:
        ax.text(0.02, 0.02, f"{hsi_labels[j]}: {hsi_vals[j]:.3f}",
                transform=ax.transAxes, ha="left", va="bottom", fontsize=FONT_MIN,
                bbox=dict(facecolor="white", alpha=0.75, edgecolor="none", pad=1.5))

handles_hsi = [Patch(facecolor='darkred', edgecolor='k', label="Highlight Area"), Patch(facecolor='deepskyblue', edgecolor='k', label="Suppressed Area")]
axes[3, 0].legend(handles=handles_hsi, loc='lower left', bbox_to_anchor=(0, 0), borderaxespad=0, fontsize=FONT_MIN, frameon=True).get_frame().set_alpha(0.85)

# ------------------------ Row 5: TGI ------------------------
tgi_vals = [TGI_shadow_Relief_vs_H300, TGI_shadow_Relief_vs_MDOW, TGI_high_Relief_vs_H300]
tgi_labels = ["TGI (SR−H)", "TGI (SR−MDOW)", "TGI (SR−H)"]
tgi_arrays = [dT_shadow_map_H300, dT_shadow_map_MDOW, dT_high_map]
img_tgi = None

for j in range(3):
    ax = axes[4, j]
    ax.imshow(np.ma.masked_invalid(poisson_ext), cmap="gray", vmin=0, vmax=1, extent=(xmin_ext, xmax_ext, ymin_ext, ymax_ext), origin="upper")
    img_tgi = ax.imshow(
        np.ma.masked_invalid(tgi_arrays[j]), cmap="bwr", vmin=-3, vmax=3,
        extent=(xmin_clip, xmax_clip, ymin_clip, ymax_clip), origin="upper", alpha=0.8
    )
    add_map_decoration(ax, "", pixel_size, x_left, x_right, y_bottom, y_top, icedoline_base)

    ax.text(0.02, 0.02, f"{tgi_labels[j]}: {tgi_vals[j]:.3f}",
            transform=ax.transAxes, ha="left", va="bottom", fontsize=FONT_MIN,
            bbox=dict(facecolor="white", alpha=0.75, edgecolor="none", pad=1.5))

# ==============================================================================
# Add upper-right panel labels such as (a), (b), (c) ...
# ==============================================================================
letters = string.ascii_lowercase  # Generate 'a', 'b', 'c', ...
for i, ax in enumerate(axes.flatten()):
    # Position: upper-right corner
    # Style: bold text with a semi-transparent white background
    ax.text(
        0.98, 0.98, f"({letters[i]})",  # Display as (a), (b), ...
        transform=ax.transAxes,
        ha="right", va="top",
        fontsize=FONT_TITLE+1,
        #fontweight="bold",
        bbox=dict(facecolor="white", edgecolor="none", alpha=0.7, pad=2)
    )

# ---------------- Layout and colorbars ----------------
plt.tight_layout(rect=[0.02, 0.02, 0.93, 0.96])
plt.subplots_adjust(wspace=0.005, hspace=0.2)

# Gray-value colorbar
cax_gray = inset_axes(axes[0, 2], width="6%", height="100%", loc="lower left", bbox_to_anchor=(1.02, 0.0, 1, 1), bbox_transform=axes[0, 2].transAxes, borderpad=0.0)
cb_gray = fig.colorbar(im_gray, cax=cax_gray)
cb_gray.ax.tick_params(labelsize=FONT_MIN)

# TGI Colorbar
cax_tgi = inset_axes(axes[4, 2], width="6%", height="100%", loc="lower left", bbox_to_anchor=(1.02, 0.0, 1, 1), bbox_transform=axes[4, 2].transAxes, borderpad=0.0)
cb_tgi = fig.colorbar(img_tgi, cax=cax_tgi, ticks=[-3, -2, -1, 0, 1, 2, 3])
cb_tgi.set_label("Δ local texture", fontsize=FONT_MIN)
cb_tgi.ax.tick_params(labelsize=FONT_MIN)

# ===== Save before calling show =====
out_path = r"E:\software_process\Python\Mapping_Swiss\process\3_reliefAnalysis\Shading_metrics_5x3.png"
fig.savefig(
    out_path,
    dpi=800,            # 800 dpi
    transparent=True,   # Transparent background
    bbox_inches="tight",
    pad_inches=0.02
)

plt.show()
```
