```python
from pathlib import Path

import numpy as np
import geopandas as gpd
import rasterio
from rasterio.mask import mask
```

# Load the input rasters and study-area shapefiles

```python
# ======================
# 1. Input paths
# ======================

stage_dir = Path.cwd().resolve()
relief_root = stage_dir.parent
data_dir = relief_root / "shared_data"

# Update these raw raster paths if the original hillshade products are moved.
input_rasters = {
    "H300": Path(r"E:\software_process\matlab_process\研一\制图论文\Hillshading\Process_possion_edit\DATA_Hillshade\LIC15_300_45_2_1m.tif"),
    "H180": Path(r"E:\software_process\matlab_process\研一\制图论文\Hillshading\Process_possion_edit\DATA_Hillshade\LIC15_180_45_2_1m.tif"),
    "H60": Path(r"E:\software_process\matlab_process\研一\制图论文\Hillshading\Process_possion_edit\DATA_Hillshade\LIC15_60_45_2_1m.tif"),
    "Relief": Path(r"E:\software_process\matlab_process\研一\制图论文\Hillshading\Process_possion_edit\Poisson_results_1m_1128\Poisson_300_60_60_norm.tif"),
    "MDOW": data_dir / "MDOW.tif",
}

study_areas = {
    "TotalArea": data_dir / "shp" / "TotalArea.shp",
    "ridge": data_dir / "shp" / "ridgeArea.shp",
    "valley": data_dir / "shp" / "valleyArea.shp",
    "icedoline": data_dir / "shp" / "IceDoline.shp",
}
```

# Create the output folders

```python
# ======================
# 2. Output folders
# ======================

clip_root = stage_dir / "intermediate_clips"
clip_root.mkdir(parents=True, exist_ok=True)

region_output_dirs = {
    "ridge": clip_root / "ridge",
    "valley": clip_root / "valley",
    "icedoline": clip_root / "icedoline",
}

for output_dir in region_output_dirs.values():
    output_dir.mkdir(parents=True, exist_ok=True)
```

# Define a reusable clipping helper

```python
# ======================
# 3. Helper function
# ======================

def clip_raster_to_shape(
    in_raster,
    shp_path,
    out_raster,
    nodata_in=-99,
    nodata_out=-9999,
    normalize_to_01=True,
):
    """
    Clip a raster by a shapefile and optionally normalize valid pixels to [0, 1].
    """
    with rasterio.open(in_raster) as src:
        gdf = gpd.read_file(shp_path)
        if gdf.crs is None:
            raise ValueError(f"Shapefile {shp_path} does not define a CRS.")
        if gdf.crs != src.crs:
            gdf = gdf.to_crs(src.crs)

        geometries = [geom for geom in gdf.geometry if geom is not None]
        out_image, out_transform = mask(src, geometries, crop=True)
        data = out_image[0].astype("float32")

        if nodata_in is not None:
            data[data == nodata_in] = np.nan

        if normalize_to_01:
            valid_mask = np.isfinite(data)
            if np.any(valid_mask):
                vmin = np.nanmin(data)
                vmax = np.nanmax(data)
                if np.isfinite(vmin) and np.isfinite(vmax) and vmax > vmin:
                    data = (data - vmin) / (vmax - vmin)

        data_out = np.where(np.isnan(data), nodata_out, data).astype("float32")

        meta = src.meta.copy()
        meta.update(
            {
                "height": data_out.shape[0],
                "width": data_out.shape[1],
                "transform": out_transform,
                "dtype": "float32",
                "nodata": nodata_out,
            }
        )

        with rasterio.open(out_raster, "w", **meta) as dst:
            dst.write(data_out, 1)

    print(f"Saved clipped raster: {out_raster}")
```

# Clip the full study area first

```python
# ======================
# 4. Step 1: clip the full study area
# ======================

total_area_outputs = {}

for product_name, raster_path in input_rasters.items():
    out_path = clip_root / f"{product_name}_TotalArea.tif"
    clip_raster_to_shape(
        in_raster=raster_path,
        shp_path=study_areas["TotalArea"],
        out_raster=out_path,
        nodata_in=-99,
        nodata_out=-9999,
        normalize_to_01=True,
    )
    total_area_outputs[product_name] = out_path
```

# Clip the three representative local areas

```python
# ======================
# 5. Step 2: clip the three representative local areas
# ======================

for region_name, shp_path in study_areas.items():
    if region_name == "TotalArea":
        continue

    output_dir = region_output_dirs[region_name]

    for product_name, raster_path in total_area_outputs.items():
        out_path = output_dir / f"{product_name}_{region_name}.tif"
        clip_raster_to_shape(
            in_raster=raster_path,
            shp_path=shp_path,
            out_raster=out_path,
            nodata_in=-9999,
            nodata_out=-9999,
            normalize_to_01=False,
        )

print(
    "Finished all representative-area clips. You can now continue with the histogram, "
    "entropy, and contrast analyses."
)
```
