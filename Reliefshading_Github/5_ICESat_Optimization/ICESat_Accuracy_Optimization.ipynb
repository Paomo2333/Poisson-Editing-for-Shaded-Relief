# ICESat Accuracy Optimization

Standalone ICESat-2 accuracy optimization workflow.

This notebook is adapted from the original local workflow but now only depends on `optimization_utils.py` in this folder.


```python
from pathlib import Path

import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

from optimization_utils import (
    DemConfig,
    ProfileColumn,
    atl06_csv_to_geodataframe,
    build_profile_dataframe,
    clip_dem_and_save,
    clip_points_to_area,
    compute_metrics,
    create_profile_figure,
    get_strong_beams_from_h5,
    load_area_of_interest,
    load_atl06_csv,
    read_dem_metadata,
    sample_dem_values,
    transform_points_for_sampling,
)

mpl.rcParams['font.family'] = 'Arial'
```

## Step 1. Path configuration

Edit this cell first if your local paths are different.


```python
base_dir = Path(r'E:\software_process\Python\Mapping_Swiss\Reviewer_Response')

atl06_csv_path = base_dir / 'Batch_Outputs' / 'CSV' / 'ATL06_20200220222523_08580610_007_01_filtered.csv'
atl06_h5_path = base_dir / 'Initial_Data' / 'ATL06_20200220222523_08580610_007_01.h5'
initial_dem_path = base_dir / 'Process_data_DEM' / 'initial_dem_new.tif'
optimized_dem_path = base_dir / 'Process_data_DEM' / 'Optimized_dem_new.tif'
rema_dem_path = base_dir / 'Process_data_DEM' / 'REMA2m.tif'
area_path = base_dir / 'SHP' / 'Area_Larsemann.shp'
output_dir = base_dir / '5_ICESat_Optimization' / 'outputs_icesat_accuracy_optimization'
output_dir.mkdir(parents=True, exist_ok=True)

print('ATL06 CSV:', atl06_csv_path)
print('ATL06 H5:', atl06_h5_path)
print('Initial DEM:', initial_dem_path)
print('Optimized DEM:', optimized_dem_path)
print('REMA DEM:', rema_dem_path)
print('Area SHP:', area_path)
print('Output dir:', output_dir)
```

```python
dem_configs = [
    DemConfig('Initial DEM', initial_dem_path, 'initial_dem_z', replace_value=-99.0),
    DemConfig('Optimized DEM', optimized_dem_path, 'optimized_dem_z', replace_value=-99.0),
    DemConfig('REMA DEM', rema_dem_path, 'rema_dem_z', replace_value=None),
]

dem_metadata = [read_dem_metadata(config) for config in dem_configs]
initial_dem_meta = dem_metadata[0]
area_gdf = load_area_of_interest(area_path, target_crs=initial_dem_meta.crs)

display(pd.DataFrame([
    {
        'dataset': meta.config.label,
        'path': str(meta.config.path),
        'crs': str(meta.crs),
        'bounds': meta.bounds,
        'resolution': meta.resolution,
    }
    for meta in dem_metadata
]))
```

```python
sc_orient, strong_beams = get_strong_beams_from_h5(atl06_h5_path)
atl06_df = load_atl06_csv(atl06_csv_path)
strong_atl06_df = atl06_df.loc[atl06_df['beam'].isin(strong_beams)].copy()
atl06_points_gdf = atl06_csv_to_geodataframe(strong_atl06_df, initial_dem_meta.crs)

orientation_label = 'forward' if sc_orient == 1 else 'backward'
print(f'sc_orient: {sc_orient} ({orientation_label})')
print(f'Strong beams: {strong_beams}')
print(f'Points before strong-beam filter: {len(atl06_df)}')
print(f'Points after strong-beam filter: {len(atl06_points_gdf)}')
display(atl06_points_gdf.drop(columns='geometry').head(12))
```

```python
clipped_atl06_gdf = clip_points_to_area(atl06_points_gdf, area_gdf)
if clipped_atl06_gdf.empty:
    raise RuntimeError('Clipped strong ATL06 is empty.')

clipped_atl06_df = clipped_atl06_gdf.drop(columns='geometry').copy()
clipped_atl06_csv_path = output_dir / f'{atl06_csv_path.stem}_strong_only_clipped.csv'
clipped_atl06_df.to_csv(clipped_atl06_csv_path, index=False, encoding='utf-8-sig')

print(f'Clipped strong ATL06 points: {len(clipped_atl06_gdf)}')
print('Saved clipped ATL06 CSV:', clipped_atl06_csv_path)
display(clipped_atl06_df.head(12))
```

```python
clip_rows = [clip_dem_and_save(meta.config, area_gdf, output_dir) for meta in dem_metadata]
clip_df = pd.DataFrame(clip_rows)
display(clip_df)
```

```python
profile_track_df = build_profile_dataframe(clipped_atl06_df.copy())
start_row = profile_track_df.sort_values('cumulative_distance_m').iloc[0]
track_map_png_path = output_dir / f'{atl06_csv_path.stem}_strong_track_map.png'

fig, ax = plt.subplots(figsize=(8, 8))
area_gdf.plot(ax=ax, facecolor='#d9dfd2', edgecolor='#7f8977', linewidth=1.0, zorder=1)
ax.scatter(
    clipped_atl06_gdf.geometry.x,
    clipped_atl06_gdf.geometry.y,
    s=12,
    color='#204a87',
    label='ICESat-2 track',
    alpha=0.85,
    linewidths=0,
    zorder=3,
    rasterized=True,
)
ax.scatter(
    start_row['x'],
    start_row['y'],
    s=220,
    marker='*',
    color='#d95f02',
    edgecolors='black',
    linewidths=0.8,
    label='Start point',
    zorder=4,
)
ax.annotate('Profile start', xy=(start_row['x'], start_row['y']), xytext=(8, 8), textcoords='offset points', fontsize=13)
min_x, min_y, max_x, max_y = area_gdf.total_bounds
ax.set_xlim(min_x, max_x)
ax.set_ylim(min_y, max_y)
ax.set_title('ICESat-2 ATL06 Track Distribution', fontsize=14)
ax.set_xlabel('X (m)', fontsize=13)
ax.set_ylabel('Y (m)', fontsize=13)
ax.set_aspect('equal')
ax.grid(True, linestyle='--', linewidth=0.5, alpha=0.6)
ax.tick_params(axis='both', labelsize=11)
ax.legend(loc='best', frameon=True, fontsize=11)
plt.show()
fig.savefig(track_map_png_path, dpi=300, bbox_inches='tight')
print('Saved track map:', track_map_png_path)
```

```python
comparison_df = clipped_atl06_gdf.drop(columns='geometry').copy()
for meta in dem_metadata:
    sample_coords = transform_points_for_sampling(clipped_atl06_gdf, meta.crs)
    comparison_df[meta.config.sample_column] = sample_dem_values(meta.config, sample_coords)

comparison_csv_path = output_dir / f'{atl06_csv_path.stem}_comparison.csv'
comparison_df.to_csv(comparison_csv_path, index=False, encoding='utf-8-sig')
print('Saved comparison CSV:', comparison_csv_path)
display(comparison_df.head(12))
```

```python
profile_df = build_profile_dataframe(comparison_df)
profile_figure = create_profile_figure(
    profile_df,
    [
        ProfileColumn('ICESat-2', 'h_li'),
        ProfileColumn('Initial DEM', 'initial_dem_z'),
        ProfileColumn('Optimized DEM', 'optimized_dem_z'),
    ],
)
ax = profile_figure.axes[0]
ax.set_title('')
ax.tick_params(axis='both', labelsize=11)
profile_png_path = output_dir / f'{atl06_csv_path.stem}_profile.png'
display(profile_figure)
profile_figure.savefig(profile_png_path, dpi=600, bbox_inches='tight')
print('Saved profile PNG:', profile_png_path)
```

```python
initial_metrics = compute_metrics(comparison_df['h_li'], comparison_df['initial_dem_z'])
optimized_metrics = compute_metrics(comparison_df['h_li'], comparison_df['optimized_dem_z'])
comparison_df['final_dem_z'] = comparison_df['optimized_dem_z'] + abs(optimized_metrics['Bias'])
final_metrics = compute_metrics(comparison_df['h_li'], comparison_df['final_dem_z'])

metrics_df = pd.DataFrame([
    {'Dataset': 'Initial DEM', **initial_metrics},
    {'Dataset': 'Optimized DEM', **optimized_metrics},
    {'Dataset': 'Final DEM', **final_metrics},
])
metrics_csv_path = output_dir / f'{atl06_csv_path.stem}_metrics.csv'
metrics_df.to_csv(metrics_csv_path, index=False, encoding='utf-8-sig')
comparison_with_final_csv_path = output_dir / f'{atl06_csv_path.stem}_comparison_with_final.csv'
comparison_df.to_csv(comparison_with_final_csv_path, index=False, encoding='utf-8-sig')
display(metrics_df)
```

```python
profile_df_final = build_profile_dataframe(comparison_df)
profile_figure_final = create_profile_figure(
    profile_df_final,
    [
        ProfileColumn('ICESat-2', 'h_li'),
        ProfileColumn('Initial DEM', 'initial_dem_z'),
        ProfileColumn('Optimized DEM', 'optimized_dem_z'),
        ProfileColumn('Final DEM', 'final_dem_z'),
    ],
)
ax = profile_figure_final.axes[0]
ax.set_title('')
ax.tick_params(axis='both', labelsize=11)
stats_text = (
    f"Initial DEM\nBias = {initial_metrics['Bias']:.2f}\nMAE = {initial_metrics['MAE']:.2f}\nRMSE = {initial_metrics['RMSE']:.2f}\nSTD = {initial_metrics['STD']:.2f}\n\n"
    f"Optimized DEM\nBias = {optimized_metrics['Bias']:.2f}\nMAE = {optimized_metrics['MAE']:.2f}\nRMSE = {optimized_metrics['RMSE']:.2f}\nSTD = {optimized_metrics['STD']:.2f}\n\n"
    f"Final DEM\nBias = {final_metrics['Bias']:.2f}\nMAE = {final_metrics['MAE']:.2f}\nRMSE = {final_metrics['RMSE']:.2f}\nSTD = {final_metrics['STD']:.2f}"
)
ax.text(0.98, 0.02, stats_text, transform=ax.transAxes, ha='right', va='bottom', fontsize=10,
        bbox=dict(boxstyle='round,pad=0.4', facecolor='white', edgecolor='gray', alpha=0.85))
profile_final_png_path = output_dir / f'{atl06_csv_path.stem}_profile_with_final.png'
display(profile_figure_final)
profile_figure_final.savefig(profile_final_png_path, dpi=600, bbox_inches='tight')
print('Saved final profile PNG:', profile_final_png_path)
```

```python
output_summary_df = pd.DataFrame([
    {'output': 'clipped_atl06_csv', 'path': str(clipped_atl06_csv_path)},
    {'output': 'comparison_csv', 'path': str(comparison_csv_path)},
    {'output': 'comparison_with_final_csv', 'path': str(comparison_with_final_csv_path)},
    {'output': 'metrics_csv', 'path': str(metrics_csv_path)},
    {'output': 'track_map_png', 'path': str(track_map_png_path)},
    {'output': 'profile_png', 'path': str(profile_png_path)},
    {'output': 'profile_with_final_png', 'path': str(profile_final_png_path)},
    *[
        {'output': row['dataset'], 'path': row['output_path']}
        for row in clip_rows
    ],
])
display(output_summary_df)
```
