# REMA_DEM_Check

Standalone quick-look notebook for `REMA_DEM.tif`.

This notebook reads the raster, converts `-99` to `NaN`, computes summary statistics, and plots a 1x2 figure.


```python
from pathlib import Path

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display

from optimization_utils import read_raster_with_nan

mpl.rcParams['font.family'] = 'Arial'

FIG_WIDTH_CM = 18.0
FIG_HEIGHT_CM = 8.5
FIGSIZE = (FIG_WIDTH_CM / 2.54, FIG_HEIGHT_CM / 2.54)
```

## Step 1. Path configuration

Edit this cell first if your local path is different.


```python
base_dir = Path(r'E:\software_process\Python\Mapping_Swiss\Reviewer_Response')
dem_path = base_dir / 'error_check' / 'REMA_DEM.tif'
output_dir = base_dir / '5_ICESat_Optimization' / 'outputs_rema_check'
output_dir.mkdir(parents=True, exist_ok=True)

print('DEM path:', dem_path)
print('Output dir:', output_dir)
```

```python
raster_info = read_raster_with_nan(dem_path)
dem = raster_info['array']
extent = raster_info['extent']
valid_values = dem[np.isfinite(dem)]

stats_df = pd.DataFrame([
    {'Metric': 'Count_valid', 'Value': int(valid_values.size)},
    {'Metric': 'Count_nan', 'Value': int(np.isnan(dem).sum())},
    {'Metric': 'Min', 'Value': float(np.nanmin(dem))},
    {'Metric': 'Max', 'Value': float(np.nanmax(dem))},
    {'Metric': 'Mean', 'Value': float(np.nanmean(dem))},
    {'Metric': 'Median', 'Value': float(np.nanmedian(dem))},
    {'Metric': 'Std', 'Value': float(np.nanstd(dem))},
    {'Metric': 'P05', 'Value': float(np.nanpercentile(dem, 5))},
    {'Metric': 'P95', 'Value': float(np.nanpercentile(dem, 95))},
])

stats_csv_path = output_dir / 'REMA_DEM_stats.csv'
stats_df.to_csv(stats_csv_path, index=False, encoding='utf-8-sig')
display(stats_df)
```

```python
mean_val = float(np.nanmean(dem))
median_val = float(np.nanmedian(dem))
std_val = float(np.nanstd(dem))
mae_val = float(np.nanmean(np.abs(dem)))
bias_val = mean_val
output_png_path = output_dir / 'REMA_DEM_difference_map_hist.png'

fig, axes = plt.subplots(1, 2, figsize=(9.2, 3.8), gridspec_kw={'width_ratios': [1, 1.45]})
axes[0].set_facecolor('#eeeeee')
im = axes[0].imshow(dem, extent=extent, origin='upper', cmap='RdBu_r', vmin=-5, vmax=5)
axes[0].set_title('Elevation Difference (REMA-Final DEM)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('X (m)', fontsize=12)
axes[0].set_ylabel('Y (m)', fontsize=12)
axes[0].tick_params(labelsize=10)
cbar = fig.colorbar(im, ax=axes[0], fraction=0.046, pad=0.04)
cbar.set_label('Elevation Difference (m)', fontsize=12)
cbar.ax.tick_params(labelsize=10)

axes[1].hist(valid_values, bins=150, color='#88a0b6', edgecolor='white', alpha=0.9)
axes[1].axvline(mean_val, color='#c44e52', linewidth=1.6, linestyle='-', label=f'Mean = {mean_val:.2f}')
axes[1].axvline(median_val, color='#4c72b0', linewidth=1.6, linestyle='--', label=f'Median = {median_val:.2f}')
axes[1].axvline(mean_val - std_val, color='#55a868', linewidth=1.2, linestyle=':', label=f'Mean - STD = {mean_val - std_val:.2f}')
axes[1].axvline(mean_val + std_val, color='#55a868', linewidth=1.2, linestyle=':', label=f'Mean + STD = {mean_val + std_val:.2f}')
axes[1].set_title('Difference Histogram', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Elevation Difference (m)', fontsize=12)
axes[1].set_ylabel('Frequency', fontsize=12)
axes[1].set_xlim(-6, 6)
axes[1].tick_params(labelsize=10)
axes[1].grid(True, linestyle='--', linewidth=0.6, color='gray', alpha=0.6)
axes[1].legend(fontsize=9, frameon=True)
axes[1].text(0.98, 0.95, f'Bias = {bias_val:.2f}\nMAE = {mae_val:.2f}', transform=axes[1].transAxes,
             ha='right', va='top', fontsize=10,
             bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='gray', alpha=0.85))
plt.tight_layout()
fig.savefig(output_png_path, dpi=600, bbox_inches='tight', transparent=True)
plt.show()
print('Saved PNG:', output_png_path)
print('Saved stats CSV:', stats_csv_path)
```
